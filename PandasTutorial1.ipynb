### APPLY with Multiple Columns

```python
#Converting a function to Lambda Expression
def simple(num):
    return num * 2
```

```python
simple(2)
```

```python
#To convert the same function to lambda function. Lambda takes in a value and return some operations on that.
#Not every function can be converted to lambda expression. Using lambda for our apply function
lambda num: num * 2
```

```python
#APPLY WITH MULTIPLE COLUMNS
df.head()
```

```python
def quality_tip(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Bad tip"
```

```python
quality_tip(16.99, 1.01)
```

```python
#Applying the function now
#First pick the columns u wanna use for your function
#Call lambda on df being passed in
#call your custom function and pass in the used columns, and specify your axis
df[['total_bill', 'tip']].apply(lambda df: quality_tip(df['total_bill'], df['tip']), axis = 1)
```

```python
#Assigning it to a column now.
df['Quality_Tip'] = df[['total_bill', 'tip']].apply(lambda df: quality_tip(df['total_bill'], df['tip']), axis = 1) 
```

```python
df.head()
```

```python
#Alternatively the above can be done through vectorization using numpy 
#Vectorization is good when the custom function is not aware that it is going to be applied to a Numpy array. Kinda better
#to use
df['Quality_Tip'] = np.vectorize(quality_tip) (df['total_bill'], df['tip'])
```

### DF Inner Merge

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
#Docs on merge
#help(pd.merge)
```

```python
pd.merge(registrations, logins, how = 'inner', on = 'name') #Using inner, we looking for something present in the two tables
```

```python
pd.merge(logins, registrations, how = 'inner', on = 'name') #The table order does not matter with inner merge
```

## Left and Right merge

```python
#IT IS IMPORTANT TO NOTE THAT THE ORDER OF THE TABLES PASSED IN AS ARGUMENT MATTERS HERE


pd.merge(registrations, logins, how = 'left', on = 'name') #Every name on our registration table which is assumed to be 
                                                        # on the left must be present
```

```python
pd.merge(registrations, logins, how = 'right', on = 'name')
                                                         #Every name on our login table which is assumed to be 
                                                        # on the right must be present
```

```python
pd.merge(left = registrations, right = logins, how = 'left', on = 'name') #Alternatively and better for the left
```

```python
pd.merge(left = registrations, right = logins, how = 'right', on = 'name') #Alternatively and better for the right
```
